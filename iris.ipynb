```python
# import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
import math

file = np.genfromtxt('train.csv', dtype=np.float, delimiter=',', skip_header = 1, usecols=(1, 2, 3, 4, 6), encoding='utf8')
idex = np.genfromtxt('train.csv', dtype=np.str_, delimiter=',', usecols=(1, 2, 3, 4, 6), max_rows=1, encoding='UTF-8-sig')
dict = {idex[0]:file[:, 0], idex[1]:file[:, 1], idex[2]:file[:, 2], idex[3]:file[:, 3]}


for i in range(list(np.shape(file))[0]):
    if i == 0:
        data = np.array([file[i][0:4]])
        label = np.array([file[i][4]])
    else:
        data = np.append(data, [file[i][0:4]], axis=0)
        label = np.append(label, [file[i][4]], axis=0)
shape = list(np.shape(data))

#print(label)
for x in range(shape[1]):
    num = 0
    for i in data[:, x]:
        if not np.isnan(i):
            num = num + i        
    if x == 0:    
        average_sepal_len = num / shape[0]
    elif x == 1:
        average_sepal_wid = num / shape[0]
    elif x == 2:
        average_petal_len = num / shape[0]
    elif x == 3:
        average_petal_wid = num / shape[0]

        
average = np.array([average_sepal_len, average_sepal_wid, average_petal_len, average_petal_wid])



#temp = data.flatten()
#print(np.shape(average))

for i in range(shape[0]):
    for j in range(shape[1]):
        if np.isnan(data[i][j]):
            for k in range(list(np.shape(average))[0]):
                if j == k:
                    data[i][j] = average[k]
#dict = {idex[0]:data[:, 0], idex[1]:data[:, 1], idex[2]:data[:, 2], idex[3]:data[:, 3]}
 

for x in range(shape[1]):
    num = 0
    for i in data[:, x]:
        if not np.isnan(i):
            num = num + (i - average[x])**2
            
    if x == 0:    
        sd_sepal_len = math.sqrt(num / shape[0])
    elif x == 1:
        sd_sepal_wid = math.sqrt(num / shape[0])
    elif x == 2:
        sd_petal_len = math.sqrt(num / shape[0])
    elif x == 3:
        sd_petal_wid = math.sqrt(num / shape[0])

sd = np.array([sd_sepal_len, sd_sepal_wid, sd_petal_len, sd_petal_wid])
ax = plt.subplot(111, frame_on=False) 
ax.xaxis.set_visible(False) 
ax.yaxis.set_visible(False) 

a = plt.table(cellText=[average, sd], rowLabels=['mean', 'sd'], colLabels=idex[0:4], loc='center') 
#a.auto_set_font_size(False)
a.set_fontsize(18)
a.scale(2.5, 2.5)
#
```
